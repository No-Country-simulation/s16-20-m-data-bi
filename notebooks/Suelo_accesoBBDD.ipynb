```python
headers = {'Content-type': 'content_type_value'}
array_coordenadas = [
    {
        'ciudad': 'Mendoza Capital',
        'coordenada': {'lon': '-68.8167', 'lat': '-32.8833'}
    },
    {
        'ciudad': 'San Rafael',
        'coordenada': {'lon': '-68.3333', 'lat': '-34.6000'}
    },
    {
        'ciudad': 'Godoy Cruz',
        'coordenada': {'lon': '-68.8333', 'lat': '-32.9167'}
    },
    {
        'ciudad': 'Tupungato',
        'coordenada': {'lon': '-69.1484', 'lat': '-33.3715'}
    },
    {
        'ciudad': 'Junín',
        'coordenada': {'lon': '-68.4667', 'lat': '-33.1333'}
    },
    {
        'ciudad': 'Tunuyán',
        'coordenada': {'lon': '-69.0167', 'lat': '-33.5667'}
    },
    {
        'ciudad': 'Luján de Cuyo',
        'coordenada': {'lon': '-68.8730', 'lat': '-32.9980'}
    },
    {
        'ciudad': 'San Martín',
        'coordenada': {'lon': '-68.4706', 'lat': '-33.0806'}
    },
    {
        'ciudad': 'Rivadavia',
        'coordenada': {'lon': '-68.4667', 'lat': '-33.1833'}
    },
    {
        'ciudad': 'Maipú',
        'coordenada': {'lon': '-68.7500', 'lat': '-32.9667'}
    },
    {
        'ciudad': 'Las Heras',
        'coordenada': {'lon': '-68.8167', 'lat': '-32.8500'}
    }
]
resultado_ciudades = np.empty(len(array_coordenadas), dtype=object)
i = 0
for coordenada in array_coordenadas:
    v_lon = coordenada['coordenada']['lon']
    v_lat = coordenada['coordenada']['lat']

    resp_probability = requests.get('https://rest.isric.org/soilgrids/v2.0/classification/query?lon='+v_lon+'&lat='+v_lat+'&number_classes=5', headers=headers)
    resp_parametros = requests.get('https://rest.isric.org/soilgrids/v2.0/properties/query?lon='+v_lon+'&lat='+v_lat+'&property=cfvo&property=clay&property=nitrogen&property=ocd&property=ocs&property=phh2o&property=sand&property=silt&property=soc&property=wv0010&property=wv0033&property=wv1500&property=bdod&property=cec&depth=0-5cm&depth=0-30cm&depth=5-15cm&depth=15-30cm&depth=30-60cm&depth=60-100cm&depth=100-200cm&value=Q0.5&value=Q0.05&value=Q0.95&value=mean&value=uncertainty', headers=headers)
    data_suelos = json.loads(resp_probability.text)
    data_parametros = json.loads(resp_parametros.text)
    resultado_json = {}
    coordenada['probability'] = data_suelos['wrb_class_probability']
    coordenada['parametros'] = data_parametros['properties']
    
    resultado_ciudades[i] = coordenada
    i+=1
```
